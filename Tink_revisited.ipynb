## Read the data

```python
data = pd.read_csv('tink.csv', encoding = "ISO-8859-1")
data.head()
```

```python
data['CLASS'].value_counts()
```

Enter the rest of the code below.

```python
inputs = data['CONTENT']
target = data['CLASS']
from sklearn.feature_extraction.text import CountVectorizer
CountVectorizer=CountVectorizer()
CountVectorizer.fit_transform(inputs)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs,target , train_size=0.8,random_state=365
                                                     ,stratify=target)
x_train_transform = CountVectorizer.fit_transform(x_train)
x_test_transform = CountVectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_transform,y_train)
y_pred=clf.predict(x_test_transform)
from sklearn.metrics import ConfusionMatrixDisplay 
ConfusionMatrixDisplay.from_predictions(
    y_test , y_pred 
)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.naive_bayes import ComplementNB
clf = ComplementNB()
clf.fit(x_train_transform,y_train)
y_pred=clf.predict(x_test_transform)
from sklearn.metrics import ConfusionMatrixDisplay 
ConfusionMatrixDisplay.from_predictions(
    y_test , y_pred 
)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
